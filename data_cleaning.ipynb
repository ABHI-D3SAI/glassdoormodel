```python
import pandas as pd
```

```python
df = pd.read_csv('Data_science_jobs.csv')
```

```python
df.head(5)
```

```python
df = df.drop(['Unnamed: 0'],axis=1)
```

```python
#df = df.drop(df.index[:30],axis=0)
```

```python
df.head(5)
```

```python
df = df.drop(df.index[0:30])
```

```python
df.reset_index()
```

```python
df = df.reset_index()
```

```python
df.head(5)
```

```python
df = df.drop(['index'],axis=1)
```

```python
df.head(5)
```

```python
df.drop(['Founded','Size','Type of Ownership','Sector','Rating','Industry'],axis=1,inplace=True)
```

```python
df = df[['Company Name','Job Title','Job Description','Salary Estimate','Location','Revenue']]
```

```python
df.head(5)
```

```python
# removing text from salary
```

```python
salary = df['Salary Estimate'].apply(lambda x: x.split('(')[0] )
```

```python
#removing ₹ and K from salary
```

```python
salary = salary.apply(lambda x: x.replace('₹','').replace('K',''))
salary = salary.apply(lambda x: x.replace('1M','1000'))
```

```python
salary.head(5)
```

```python
# getting min_salary and max_salary
```

```python
salary = salary.apply(lambda x:x.replace(' ',''))
```

```python
df['min_salary'] = salary.apply(lambda x:int(x.split('-')[0]))
df['max_salary'] = salary.apply(lambda x:int(x.split('-')[-1]))
```

```python
# checking rrquirenments as python,sql and machine learning, tensoflow
```

```python
df['python'] = df['Job Description'].apply(lambda x: 1 if 'python' in x.lower() else 0)
```

```python
df['sql'] = df['Job Description'].apply(lambda x: 1 if 'sql' in x.lower() else 0)
```

```python
df['Machine Learning'] = df['Job Description'].apply(lambda x: 1 if 'machine learning' in x.lower() else 0)
```

```python
df['TensorFlow'] = df['Job Description'].apply(lambda x: 1 if 'tensorflow' in x.lower() else 0)
```

```python
df.to_csv('data_clean.csv')
```
